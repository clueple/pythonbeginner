# Get Source Data And Store In a Table ('bk')

```python
'''
Import libraries
'''
import pandas as pd

'''
Store the 'data_source'(in Excel format) in a table called 'bk'
'''
bk = pd.read_html(url)[0]
bk.drop('Unnamed: 0',axis=1,inplace=True)

# print out bk (optional)
bk
```

# Sorted out the unique system invoice number from table,'bk'

```python
# uq = bk.groupby('system_inv_num').head(1)
# uq
uq = bk.groupby('system_inv_num').head(1)
uq
```

# Create a table, 'je', and return the transaction amount ('trans_amt') 
# for each system invoice number ('system_inv_num')

```python
'''
Create table,je, with bk['je_num'] as the index column
'''
je = pd.DataFrame(index=bk['je_num'], columns={'dummy1','dummy2'})
je['system_inv_num'] = je.index.map(bk.set_index('je_num')['system_inv_num'].to_dict())
je
```

# When duplicate 'system_inv_num', you cannot use map() to the original source table, 'bk'

```python
je['trans_amt'] = je['system_inv_num'].map(bk.set_index('system_inv_num')['trans_amt'].to_dict())
je
```
